```python
bg['Salary'] = bg['Salary'].apply(lambda x:x.replace("$",'')) ##replace charaters
bg['Salary'] = bg['Salary'].apply(lambda x:x.replace("k",''))
```

```python
##split values into two
def average_of_this_two(args):
    try :
        return float(args)
    except:
        values = args.split('-')
        
        if len(values) == 2:
            
            return float(int(values[0]) + int(values[1]) ) /2
        
bg['Salary'] = bg["Salary"].apply(average_of_this_two)
```
